```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_file = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_file = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_df = pd.merge(mouse_file, clinical_trial_file,
                                 how='outer', on='Mouse ID')

# Display the data table for preview
combined_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped = combined_df.groupby(["Drug","Timepoint"])

mean_tumor = grouped.mean()["Tumor Volume (mm3)"]

# Convert to DataFrame
avg_t_vol = pd.DataFrame(mean_tumor) 

# Preview DataFrame
avg_t_vol.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem_tumor = grouped.sem()["Tumor Volume (mm3)"]

# Convert to DataFrame
sem_t_vol = pd.DataFrame(sem_tumor) 

# Preview DataFrame
sem_t_vol.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
table_mean=pd.pivot_table(avg_t_vol, values = ('Tumor Volume (mm3)'), index=['Timepoint'], columns=['Drug'], aggfunc=np.mean)

# Preview that Reformatting worked
table_mean
                                   
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met = grouped.mean()["Metastatic Sites"]

# Convert to DataFrame
avg_met = pd.DataFrame(mean_met) 

# Preview DataFrame
avg_met.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
sem_met = grouped.sem()["Metastatic Sites"]

# Convert to DataFrame
sem_met = pd.DataFrame(sem_met) 

# Preview DataFrame
sem_met.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
met_resp_table=pd.pivot_table(avg_met, values = ('Metastatic Sites'), index=['Timepoint'], columns=['Drug'], aggfunc=np.mean)

# Preview that Reformatting worked
met_resp_table.head(3)
```
